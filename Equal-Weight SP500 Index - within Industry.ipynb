### Implement Strategy

Industry Weighted (SIC) by industry market cap, equal weighted within industry.

```python
import numpy as np

strategyReturns = {'Period': [], 'strategyReturns': []}

years = pd.date_range('2014-01-01', '2018-01-01', freq='Y')

for year in years:
    # Get top 500 largest market cap companies for the given year
    marketCap500 = monthlySecurityInfo[(pd.DatetimeIndex(monthlySecurityInfo['mcaldt']).year==year.year) & 
                             (pd.DatetimeIndex(monthlySecurityInfo['mcaldt']).month==year.month)][['mtcap', 'kypermno']]
    
    marketCap500 = marketCap500.groupby(by='kypermno').sum().reset_index()
    marketCap500 = marketCap500.sort_values(by='mtcap', ascending=False)[:500]
    
    # Keep Only Identified Constituents
    constituentData = pd.merge(marketCap500.drop('mtcap',axis=1), monthlySecurityInfo, how='inner', left_on='kypermno', right_on='kypermno')[['kypermno', 'hcomnam', 'mret', 'mcaldt', 'mprc', 'mtcap', 'SIC_Classification']]

    months = pd.date_range(year - pd.DateOffset(months=11), year , freq='M')
    for month in months:
        constituentDataMonth = constituentData[(pd.DatetimeIndex(constituentData['mcaldt']).year==month.year) & 
                                 (pd.DatetimeIndex(constituentData['mcaldt']).month==month.month)]
        # Find Weights of Individual SIC Sectors For Given Month (mtcap weighted)
        SICWeights = constituentDataMonth[['mtcap', 'SIC_Classification']]
        SICWeights = SICWeights.groupby(by='SIC_Classification').sum().reset_index()
        SICWeights['mtcapPerc'] = SICWeights['mtcap']/SICWeights['mtcap'].sum()
        SICWeights = SICWeights.set_index('SIC_Classification')
        SICWeights = SICWeights['mtcapPerc'].to_dict()
        constituentDataMonth['Sector_Weight'] = constituentDataMonth['SIC_Classification'].map(SICWeights)
        
        # Assign Weights of companies within SIC Sectors For Given Month (equal weighted)
        internalWeights = constituentDataMonth[['mtcap', 'SIC_Classification']]
        internalWeights = 1/internalWeights['SIC_Classification'].value_counts()
        internalWeights = internalWeights.to_dict()
        constituentDataMonth['Internal_Weight'] = constituentDataMonth['SIC_Classification'].map(internalWeights)
        
        # Calculate Returns
        returns = constituentDataMonth['Sector_Weight']*constituentDataMonth['Internal_Weight']*constituentDataMonth['mret'].astype(float)

        strategyReturns['strategyReturns'].append((returns.sum()))
        strategyReturns['Period'].append(month)


currentPeriod = []
strategyValues = {'Period': strategyReturns['Period'], 'strategyValues': []}
currentValue = 100
strategyValues['strategyValues'].append(currentValue)
for return_value in strategyReturns['strategyReturns']:
    currentValue = currentValue * (1 + return_value)
    strategyValues['strategyValues'].append(currentValue)

strategyValues = {'Period': strategyReturns['Period'], 'strategyValues': strategyValues['strategyValues'][1:]}
```
